# Getting started: using the LCW-cDFT code

First you need to create an input file. An example is given in the  `example.in` file.
The first section has following format:

```

# FLUID PROPERTIES

model                       = SPC/E
bulk_density [AA-3]         = 0.03323615
dcf_kspace [file_path]      = ../parameters/dcf_ck_SPCE_T300.txt
temperature [K]             = 300
chemical_potential [kBT]    = 1e-3

```

The key parameter to note here is `dcf_kspace` where the correct (relative or absolute) path to the file storing the direct correlation function needs to be specified.



```
# SQUARE GRADIENT

c [J mol-1 nm9]             = 1.0793454634568314
d [J mol-1 nm15]            = 30
m_min [kJ mol-2 cm3 AA2]    = 870
m_max [kJ mol-2 cm3 AA2]    = 1255
rho_s_min [AA-3]            = 0.018943804425
rho_s_max [AA-3]            = 0.0
liquid_coex_density [AA-3]  = 0.033234078757543585
vapor_coex_density [AA-3]   = 4.747e-7

```

You then need to specify the parameter for the square gradient functional governing the slowly varying density field.



```

# EXTERNAL POTENTIAL

solute_type                 = HS
HS_radius [AA]              = 2

```


This part specifies that the external potential is from a hard sphere of radius 2 angstrom.



```

# COARSE-GRAINING PARAMETERS

a [kJ cm3 mol-2]            = 200
lambda [angstrom]           = 0.8

```

The coarse-graining parameter for the potential coupling the slowly varying and rapidly varying component of the density field.


Any other optional settings can also be specified in the input file.
